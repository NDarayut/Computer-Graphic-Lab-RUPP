```python
import cv2
```

```python
image = cv2.imread("./image/212657710_200x200.jpg", cv2.IMREAD_COLOR)
```

```python
cv2.imshow("Darayut", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite("darayut.jpg", image)
```

```python
print(f"Image shape: {image.shape}")
print(f"Image height: {image.shape[0]}")
print(f"Image width: {image.shape[1]}")
print(f"Image channels: {image.shape[2]}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(image) # Matplotlib uses RGB color format, but opencv uses BGR, so we have to convert BGR to RGB
```

```python
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # Converting BGR to RGB
plt.imshow(RGB_image)
```

```python
GRAY_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(GRAY_image, cmap=plt.cm.gray) # Use for plotting grayscale images
```

```python
print(image) # Printing the raw pixel value of the image
```

```python
print(f"Lowest pixel value: {image.min(axis=(0, 1))}")
print(f"Highest pixel value: {image.max(axis=(0, 1))}")
```
